Retrieves a list of all countries from an API.

Returns:
    list: A list of country codes.

```python
import time
import requests
from database_connector_countries import *

def get_all_countries_from_api():
    country_map = []
    url = "https://wft-geo-db.p.rapidapi.com/v1/geo/countries"
    headers = {
        "X-RapidAPI-Key": "bcb6679012msh169f4edcd407c6dp17800bjsnb43444ccba7a",
        "X-RapidAPI-Host": "wft-geo-db.p.rapidapi.com"
    }
    countries = []
    page = 1
    while True:
        params = {
            "offset": len(countries),
            "limit": 10  # Adjust the limit as per your requirement
        }
        response = requests.get(url, headers=headers, params=params, timeout=5)
        data = response.json()
        countries.extend(data['data'])
        if len(countries) >= data['metadata']['totalCount']:
            break
        page += 1
        time.sleep(2)
        country_map.extend([country['code'] for country in data['data']])

    return country_map
```
